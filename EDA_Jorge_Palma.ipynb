```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
```

```python
drive.mount('/content/drive/')
```

```python
data = pd.read_csv('drive/MyDrive/data/creditcard.csv')
```

```python
Class = pd.DataFrame(data['Class'],columns=['Class'])
```

```python
data.drop(['Time', 'Class'], axis=1,inplace=True)
```

```python
fig = plt.figure(figsize = (15, 20))
j = 0
for i in data.columns:
    plt.subplot(8, 4, j+1)
    j += 1
    sns.distplot(data[i][Class['Class']==0], color='g', label = 'Non Fraud')
    sns.distplot(data[i][Class['Class']==1], color='r', label = 'Fraud')
    plt.legend(loc='best')
    
fig.suptitle('Data Analysis')
fig.tight_layout()
fig.subplots_adjust(top=0.95)
plt.show()
```
